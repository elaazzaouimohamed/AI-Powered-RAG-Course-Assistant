````python
import sys
sys.path.insert(0, '../src')

from extraction.pdf_extractor import extraire_pdf

# ── Configurer le chemin vers un vrai PDF de cours
CHEMIN_PDF = '../data/pdfs/mon_cours.pdf'
DOSSIER_IMAGES = '../data/images'
````

## 1. Extraction PDF

````python
pages = extraire_pdf(CHEMIN_PDF, dossier_images=DOSSIER_IMAGES)
print(f'Nombre de pages extraites : {len(pages)}')
````

## 2. Inspection des résultats

````python
import pandas as pd

# Vue d'ensemble des pages
df = pd.DataFrame([
    {
        'page': p['page'],
        'methode': p['methode_extraction'],
        'nb_chars': len(p['text']),
        'image': p['contient_image'],
        'alerte_formule': p['alerte_formule_complexe'],
        'repetitions': len(p['est_entete_pied_page_repete']),
    }
    for p in pages
])

df
````

````python
# Afficher le texte d'une page spécifique
NUM_PAGE = 1  # modifier selon le PDF testé

page = next(p for p in pages if p['page'] == NUM_PAGE)
print(f"--- Page {page['page']} ({page['methode_extraction']}) ---")
print(page['text'])
````

````python
# Afficher l'image d'une page si elle a été sauvegardée
from IPython.display import Image as IPImage

pages_avec_image = [p for p in pages if p['image_path']]
if pages_avec_image:
    premiere_image = pages_avec_image[0]['image_path']
    print(f'Affichage : {premiere_image}')
    IPImage(filename=premiere_image, width=600)
else:
    print('Aucune page avec image trouvée.')
````
